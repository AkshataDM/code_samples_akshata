Q1 Read in data

Method - Read in data using read_csv() function and using the "\t" separator to read in the text file. 

```python
gold_df = pd.read_csv("gold.txt", sep = "\t",header = None)
labels_df = pd.read_csv("labels.txt",sep = "\t", header = None)
```

Q2 Split into two data frames

Method - Creating a mask by using the DataFrame.isin() function to check for URL that are in gold_df and labels_df

```python
mask = labels_df[1].isin(gold_df[0])
labels_on_gold = labels_df[mask]
labels_unknown = labels_df[~mask]
labels_on_gold
```

Q3 Compute accuracies of turks

Method - 
I first merge the two dataframes - labels_on_gold and gold_df on the "links" column. So, now we have for each link, all the ratings of different turks.
I create a dataframe ratings_count_df which stores the counts of ratings for each turk grouped on the "Turk ID" column. 
Then, I create a mask where "Ratings_x" and "Ratings_y" of the merged table are equal. I apply this mask on the merged_table df and this gives me a new data frame where all turks have rated the links correctly. I apply a groupby() function on this data frame and this gives me the count of correct ratigns grouped by "Turk ID" column. I divide this data frame by the ratings_count_df data frame to give me accuracies of each turk. I append "Turk ID", "Number of ratings" and "Accuracy" columns obtained in the three steps above to get me the rater_goodness data frame. 

Assumptions - I will not be removing duplicates by Turk ID, Links and Ratings because the same turk might have been given the same link multiple times to test his/her consistency. Removing it will affect the turks' accuracy if the turk has given the same rating to the link all times. 

```python
#creating column names for data frames
labels_on_gold.columns = ["Turk ID", "Links", "Ratings"]
gold_df.columns = ["Links", "Ratings"]

#merging the two data frames on the Links column - so merged_table has only those links common to both data frames
merged_table = pd.merge(labels_on_gold, gold_df, left_on = "Links", right_on = "Links")

#creating a mask where the ratings by turk and ratings on gold_df are the same
mask_of_correctness = merged_table["Ratings_x"] == merged_table["Ratings_y"]

#counting the number of ratings given by each turk using groupby
ratings_count_df = merged_table.groupby(["Turk ID"])[["Turk ID","Links","Ratings_x","Ratings_y"]].count()

#applying the mask on the merged_table and creating new df
correct_df = merged_table[mask_of_correctness]

#counting the number of times the turk is correct
accuracy_count_df = correct_df.groupby(["Turk ID"])[["Turk ID","Links","Ratings_x","Ratings_y"]].count()

#calculating the accuracy
accuracy_df = accuracy_count_df/ratings_count_df

#when the turk never gets the ratings right, his accuracy_count_df will not have values. So, in the accuracy_df, there will be NaN for this turk. We fill up this NaN value with 0 
accuracy_df = accuracy_df.fillna(0)


#creating rater_goodness dataframe 
rater_goodness = pd.DataFrame()
rater_goodness["Turk ID"] = accuracy_df["Turk ID"].index  
#when the turk is never correct, these turks do not appear in the correct_df. So, NaN values result and these are filled with 0

#Accuracy is the average correctness of the turk
rater_goodness["Accuracy"] = list(accuracy_df["Links"])
rater_goodness["Number of ratings"] = list(ratings_count_df["Links"])
rater_goodness.sort_values(by = "Number of ratings", ascending = False)
#Set the turk ID column as index
rater_goodness 
```

Q4 Odds ratios 

Method - I create a simple function to calculate the odds ratio and then create a map to apply this function to each value in the column "Accuracy" 

```python
#simple function to calculate the odds
def odds_ratio(p):
    odds = p*1.0/((1.001-p)*1.0)
    return odds

rater_goodness["Odds ratio"] = rater_goodness["Accuracy"].map(odds_ratio)
rater_goodness.sort_values(by = "Number of ratings", ascending = False)
```

Q 5 Most accurate turks 

Method - I first create a mask where the number of ratings for the turk is >= 20. After applying this mask to the rater_goodness data frame, I sort the data frame by "Accuracy" column and in descending order

```python
high_ratings_mask = rater_goodness["Number of ratings"] >= 20
high_ratings = rater_goodness[high_ratings_mask]
high_ratings.sort_values(by = "Accuracy", ascending = False)
```

Q 7 Overall predicted odds 

Method - I first set the three column names for the labels_unknown dataframe. Then, from the rater_goodness data frame, I select the value of the 75th percentile in the number of ratings column. Creating a mask with this value at the 75th percentile, I choose those rows from the data frame whose number of ratings value is greater than or equal to this value.
From the new data frame, I choose the value of turk id and odds ratio and store it in high_turk_odds_ratio. I apply a pd.merge() on the labels_unknown dataframe and the high_turk_odds_ratio data frame. Then, I create a pivot table to multiple the odds ratio of links in each category. We will get some NaN values which are replaced with 0 for better readability.

```python
labels_unknown.columns = ["Turk ID", "Links", "Ratings"]

turks_with_greater_75 = rater_goodness["Number of ratings"].quantile(q = 0.75)
mask_of_turks_greater_75 = rater_goodness["Number of ratings"] > turks_with_greater_75
high_turks = rater_goodness[mask_of_turks_greater_75]
high_turk_odds_ratio = high_turks[["Turk ID", "Odds ratio"]]
merge_high_turk_labels = pd.merge(labels_unknown,high_turk_odds_ratio,left_on = "Turk ID", right_on = "Turk ID")
overall_odds_pivot = pd.pivot_table(merge_high_turk_labels,index = "Links",columns = 'Ratings', values = "Odds ratio",aggfunc = numpy.prod)
overall_odds_pivot = overall_odds_pivot.fillna(0)
overall_odds_pivot
```

Q 8 Predicted categories

Method - I use the idxmax() function to get the index of the value that is maximum in the row ( category ). So, the top_category is a series which contains the top category label for each row. 
Then, I apply the max function to get at the maximum value and store it in a series called top_odds. 
Finally, I create a result_75 data frame with the two series as columns and column names as 'Top category' and 'Top odds' respectively.

```python
top_category = overall_odds_pivot.idxmax(axis = 1)
top_odds = overall_odds_pivot.max(axis = 1)
result_75 = pd.DataFrame({"Top category":top_category, "Top odds":top_odds})
result_75
```
